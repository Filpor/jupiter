```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
df=pd.DataFrame(prices, columns=['month','price_PLN'])
df=df.set_index('month')
df=df.assign(Price_in_USD=lambda x: x.price_PLN/4)
df['Price_in_USD'].plot(kind="line", color="red", linestyle="dashed", title="Price of goods (USD)")


```
